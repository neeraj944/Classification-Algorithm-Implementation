```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
dt_acc = accuracy_score(y_test, dt_pred)
dt_acc
#How it works:
#Splits the dataset into branches based on feature thresholds, forming a tree structure.

#Why suitable:
#Handles nonlinear relationships and does not require feature scaling.
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42, n_estimators=100)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_acc = accuracy_score(y_test, rf_pred)
rf_acc
#How it works:
#Builds multiple decision trees and averages their predictions to reduce overfitting.

#Why suitable:
#Performs well on structured data and improves accuracy through ensemble learning.
```

```python
# Support Vector Machine
from sklearn.svm import SVC
svm = SVC(kernel='rbf', random_state=42)
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)
svm_acc = accuracy_score(y_test, svm_pred)
svm_acc
# How it works:
#Finds the hyperplane that maximizes the margin between classes. The RBF kernel handles nonlinear separation.

#Why suitable:
#Excellent performance on small to medium-sized datasets with clear class separation.
```

```python
# k-Nearest Neighbors 
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
knn_acc = accuracy_score(y_test, knn_pred)
knn_acc
# How it works:
#Predicts the class based on the majority vote among the k nearest training samples.

#Why suitable:
#Simple and effective for classification when decision boundaries are irregular.
```

```python
# model comparison
results = pd.DataFrame({
    'Model': ['Logistic Regression', 'Decision Tree', 'Random Forest', 'SVM', 'k-NN'],
    'Accuracy': [lr_acc, dt_acc, rf_acc, svm_acc, knn_acc]
}).sort_values(by='Accuracy', ascending=False).reset_index(drop=True)

print(results)
```
